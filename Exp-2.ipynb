Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Line Plot
x=np.array([0,10])
y=np.array([0,500])
plt.plot(x,y)
plt.show()
```

```python
#Line plot
x=np.array([0,4,6,8])
y=np.array([400,250,450,150])
plt.plot(x,y)
plt.show()
```

```python
#Line plot
x=np.array([0,4,6,8])
y=np.array([400,250,450,150])
plt.plot(x,y,marker='o')
plt.show()
```

```python
#Line plot
x=np.array([0,4,6,8])
y=np.array([400,250,450,150])
plt.plot(x,y,marker='s')
plt.show()
```

```python
#Line plot
x=np.array([0,4,6,8])
y=np.array([400,250,450,150])
plt.plot(x,y,marker='h',ms=18,mec='g',mfc='w')
plt.show()
```

```python
#Line plot
x=np.array([0,2,4,6,8])
y=np.array([150,350,200,350,150])
plt.plot(x,y,ls='dotted',marker='h',ms=18,mec='g',mfc='w')
plt.show()
```

```python
#Line plot
x=np.array([0,2,4,6,8])
y=np.array([150,350,200,350,150])
plt.plot(x,y,ls='dotted',marker='h',ms=18,mec='g',mfc='w')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.grid()
```

```python
#lineplot
x=np.array([10,2,5,8,3])
y=np.array([1,8,4,9,2])
plt.plot(x)
plt.plot(y)
```

```python
#bar Chart
x=np.array(['A',' B','C','D'])
y=np.array([2,8,4,6])
plt.bar(x,y,width=0.5)
plt.xlabel('Letters')
plt.ylabel('Size')
plt.title('Bar Chart')
plt.show()
```

```python
#Pie Chart
x=np.array([2,4,2,2])
colors=('blue','green','pink','gray')
label=['Book','Pen','Pencil','Record']
plt.pie(x,labels=label,explode=(0.2,0,0,0),autopct='%1.2%%i',colors=colors)
plt.legend()
plt.show()
```

```python
#Histogram
x=np.random.normal(170,10,230)
plt.hist(x)
plt.show()
```

```python
#Scatter Plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
plt.show()
```
